### Matplotlib
- 그래프를 그리기위해 사용할 값들을 미리 구해놔야 한다.

```python
# 선 그래프
# 어떠한 데이터 흐름에 따라 변화되는 데이터를 표현할 때 사용한다.
x = [1, 2, 3]
y1 = [2, 3, 1]
y2 = [1, 2, 3]

# 제목
plt.title('선 그래프')

# 선 그래프를 그린다.
plt.plot(x, y1, label='선1')
plt.plot(x, y2, label='선2')

# 범례
plt.legend(bbox_to_anchor=(1.1, 1))

# 각 축의 이름
plt.xlabel('x 축')
plt.ylabel('y 축')

# 각 축의 눈금
plt.xticks([1, 2, 3], ['가', '나', '다'])
plt.yticks([0, 2, 4], ['하', '중', '상'])

plt.show()
```

- 선 그래프는 x 축 데이터의 변화에 따른 y 축 값의 변화를 해석하기 위해 사용한다
- 시간에 따른 주식 시세 변화, 시간에 따른 환율 변화 등등

```python
# 막대 그래프
x = [1, 2, 3]
y = [2, 3, 1]

plt.title('막대 그래프')
plt.bar(x, y)

# 각 축의 이름
plt.xlabel('x 축')
plt.ylabel('y 축')

# 각 축의 눈금
plt.xticks([1, 2, 3], ['가', '나', '다'])
plt.yticks([0, 2, 4], ['하', '중', '상'])

plt.show()
```

- 각 데이터의 양을 비교하는 데 사용한다.
- 범주형 데이터의 도수를 비교하고 해석하는 용도로 사용한다.

```python
# 가로 방향 막대 그래프
# 막대 그래프
x = [1, 2, 3]
y = [2, 3, 1]

plt.title('막대 그래프')
plt.barh(x, y)

# 각 축의 이름
plt.ylabel('x 축')
plt.xlabel('y 축')

# 각 축의 눈금
plt.yticks([1, 2, 3], ['가', '나', '다'])
plt.xticks([0, 2, 4], ['하', '중', '상'])

plt.show()
```

```python
# 파이 차트
# 원에서 어느 정도를 차지하는지를 표현한다.
# 데이터에 대한 도수를 비율로 표현하여 해석할 수 있는 그래프
# 파이 차트를 그리기 위한 데이터의 상대도수를 구하지 않아도 된다.
values = [15, 30, 40, 10]

a1 = ['값1', '값2', '값3', '값4']

# https://matplotlib.org/stable/gallery/color/named_colors.html
a2 = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

a3 = [0, 0.1, 0, 0]

a4 = 90

a5 = False

a6 = '%.2f%%'

a7 = True

# labels : 각 파이 조각의 라벨
# colors : 각 파이의 색상
# explode : 강조하고 싶은 파이 조각을 튀어나오게 해주는 비율, 1 = 원의 반지름
# startangle : 첫 조각의 시각 위치 각도. 0 이면 3시 방향이며 양수로 설정하면 시계 반대방향
# 음수로 설정하면 시계방으로 시작위치가 돌아간다
# counterclock : 원을 그리는 방향을 설정한다 True면 시계 반대 방향(기본), False
# 면 시계방향으로 그린다.
# autopct : 파이 조각에 보여줄 비율 수치 양식. %.2f%% - 소수점 이하 둘 째자리까지 보여주겠다는 의미
# shadow : 그림자 표시 여부 False면 그림자를 표시하지 않으며(기본), True면 그림자를 표시해준다.
plt.pie(values, labels=a1, colors=a2, explode=a3, 
        startangle=a4, counterclock=False, autopct=a6, shadow=a7)

plt.show()
```

```python
# 히스토그램
# 연속성을 가지고 있는 수치 데이터에 대한 도수 분포를 보고자 할 때 사용한다.
# 계급의 범위를 정해 각 계급별로 도수를 확인할 수 있다.
x = np.random.randn(1000)

plt.title('히스토그램')
# plt.hist(x)
# 계급의 개수를 설정하고 싶다면 bins를 사용한다. 기본이 10
# plt.hist(x, bins=30)
# 계급의 범위를 직접 설정할 수도 있다.
plt.hist(x, bins=[-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.show()
```

```python
x1 = np.random.rand(100)
y1 = np.random.rand(100)

x2 = np.random.rand(100)
y2 = np.random.rand(100)

# x축의 데이터와 y축의 데이터를 1:1 로 매칭해여 만나는 지점에 점을 찍는 그래프다
# 데이터의 분포를 확인하기 위한 용도로 사용한다.
plt.title('산점도')
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
```

### pandas를 활용한 시각화

```python
# 선 그래프
x = np.random.randn(100, 3)
# index : 2021년 1월 1일 부터 100일치 날짜를 가져온다..
index = pd.date_range('2021/01/01', periods=100)

df1 = pd.DataFrame(x, index=index, columns=['A', 'B', 'C'])
df1
```

```python
# 컬럼 하나의 데이터가 선 하나가 된다.
# index가 x 축에 적용된다.
# 컬럼의 값들은 y 축에 적용된다.
# 컬럼의 이름은 각 선의 label에 적용되며 범례에 나타난다.
df1.plot()

plt.title('선 그래프')
plt.xlabel('시간')
plt.ylabel('데이터')

plt.show()
```

### seaborn
- 데이터 프레임에서 x 축으로 사용할 컬럼과 y 축으로 사용할 컬럼, 그룹의 기준이 될 컬럼의 이름을 직접 지정할 수 있다.
- 그래프를 그리위해 구해야 하는 값을 직접 구해주기 때문에 통계 분석시 편리하게 사용할 수 있다.
- 이뻐요~

```python
# rug plot
# 1차원 데이터에 저장되어 있는 값들의 종류를 확인할 수 있다.
# 각 값들이 얼마나 저장되어 있는지는 확인할 수 없다.
sns.rugplot(data=iris, x='꽃잎길이')
plt.title('Rug Plot')
plt.show()
```

```python
# kde
sns.kdeplot(data=iris, x='꽃잎길이')
plt.title('kde plot')
plt.show()
```

```python
# kde
# x나 y에 컬럼이름을 지정하지 않으면 데이터 프레임이 가지고 있는 컬럼들 중에 그래프로 표현할 수
# 있는 모든 컬럼들의 그래프를 그려준다.
sns.kdeplot(data=iris)
plt.title('kde plot')
plt.show()
```

```python
# 하나의 좌표 평면에 히스토그램, kde, rug plot을 같이 표현한 그래프이다.
sns.displot(data=iris, x='꽃잎길이', kde=True, rug=True)
plt.show()
```

```python
# count plot : 범주형 데이터에 대한 도수를 시각화한다.
# 필요한 도수를 직접 구해준다.
sns.countplot(data=iris, x='품종')
plt.title('count plot')
plt.show()
```

```python
# 산점도와 히스토그램을 같이 표현한다.
sns.jointplot(data=iris, x='꽃받침길이', y='꽃받침너비')
plt.suptitle('joint plot', y=1.02)
plt.show()
```

```python
# jointplot을 그릴때 kind에 kde를 넣어주면 kde 밀도 그래프를 볼 수 있다.
sns.jointplot(data=iris, x='꽃받침길이', y='꽃받침너비', kind='kde')
plt.suptitle('joint plot kde', y=1.02)
plt.show()
```

```python
# 주어진 데이터 프레임에서 숫자 컬럼들만 모아서 각 컬럼의 조합으로 된 모든 산점도를 
# 보여준다. 같은 컬럼의 조합에서는 히스토그램 형태로 보여준다.
sns.pairplot(iris)
plt.suptitle('pair plot', y=1.02)
plt.show()
```

```python
# seaborn이 제공하는 대부분의 함수들은 hue 매개변수를 통해 그룹의 기준을 설정할 수 있다.
# pairplot의 경우 그룹의 기준을 설정해주면 각 그룹별 산점도를 확인할 수 있으며
# 같은 컬럼의 조합에서는 각 그룹별 kde를 보여준다.
sns.pairplot(iris, hue='품종')
plt.suptitle('pair plot', y=1.02)
plt.show()
```

```python
a1 = iris.drop('품종', axis=1).corr()
# heatmap : 값이 작은 쪽은 어둡게, 값이 큰 쪽을 밝게하여 표현하는 그래프이다.
# 각 컬럼간의 피어슨 상관관계 값을 시각화하는 용도로 많이 사용한다.
# https://seaborn.pydata.org/tutorial/color_palettes.html
sns.heatmap(a1, annot=True, cmap='summer')
plt.show()
```

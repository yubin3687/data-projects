```python
class DoubleLinkedList:
    def __init__(self):
        self.linked_list = {
            'head': [None, None, 'node_0'], 
            'tail': ['node_0', None, None]   
        }
        self.node_count = 1  
        self.linked_list['node_0'] = [None, None, 'tail']

    
    def addData(self, value, head_or_tail):
 
        if head_or_tail == 0:
            new_node_name = f'node_{self.node_count}'
            self.node_count += 1
            self.linked_list[new_node_name] = [None, value, self.linked_list['head'][2]]
            self.linked_list[self.linked_list['head'][2]][0] = new_node_name
            self.linked_list['head'][2] = new_node_name
        else:
            new_node_name = f'node_{self.node_count}'
            self.node_count += 1
            self.linked_list[new_node_name] = [self.linked_list['tail'][0], value, None]
            self.linked_list[self.linked_list['tail'][0]][2] = new_node_name
            self.linked_list['tail'][0] = new_node_name

    def showAllData(self, head_or_tail):
        # head_or_tail == 0이면 앞에서부터, 1이면 뒤에서부터
        if head_or_tail == 0:
            current_node_name = self.linked_list['head'][2]
            while current_node_name != 'tail':
                print(self.linked_list[current_node_name][1])
                current_node_name = self.linked_list[current_node_name][2]
        else:
            current_node_name = self.linked_list['tail'][0]
            while current_node_name != 'head':
                print(self.linked_list[current_node_name][1])
                current_node_name = self.linked_list[current_node_name][0]

    # 특정 위치의 데이터 가져오기
    def getData(self, position, head_or_tail):
        if head_or_tail == 0:
            current_node_name = self.linked_list['head'][2]
            current_position = 0
            while current_node_name != 'tail':
                if current_position == position:
                    return self.linked_list[current_node_name][1]
                current_node_name = self.linked_list[current_node_name][2]
                current_position += 1
        else:
            current_node_name = self.linked_list['tail'][0]
            current_position = 0
            while current_node_name != 'head':
                if current_position == position:
                    return self.linked_list[current_node_name][1]
                current_node_name = self.linked_list[current_node_name][0]
                current_position += 1

    # 특정 위치에 데이터 삽입하기
    def insertData(self, position, value, head_or_tail):
        if head_or_tail == 0:
            current_node_name = self.linked_list['head'][2]
            current_position = 0
            while current_node_name != 'tail':
                if current_position == position:
                    new_node_name = f'node_{self.node_count}'
                    self.node_count += 1
                    self.linked_list[new_node_name] = [self.linked_list[current_node_name][0], value, current_node_name]
                    self.linked_list[self.linked_list[current_node_name][0]][2] = new_node_name
                    self.linked_list[current_node_name][0] = new_node_name
                    break
                current_node_name = self.linked_list[current_node_name][2]
                current_position += 1
        else:
            current_node_name = self.linked_list['tail'][0]
            current_position = 0
            while current_node_name != 'head':
                if current_position == position:
                    new_node_name = f'node_{self.node_count}'
                    self.node_count += 1
                    self.linked_list[new_node_name] = [self.linked_list[current_node_name][0], value, current_node_name]
                    self.linked_list[self.linked_list[current_node_name][0]][2] = new_node_name
                    self.linked_list[current_node_name][0] = new_node_name
                    break
                current_node_name = self.linked_list[current_node_name][0]
                current_position += 1

    # 특정 위치의 데이터 수정하기
    def setData(self, position, value, head_or_tail):
        if head_or_tail == 0:
            current_node_name = self.linked_list['head'][2]
            current_position = 0
            while current_node_name != 'tail':
                if current_position == position:
                    self.linked_list[current_node_name][1] = value
                    break
                current_node_name = self.linked_list[current_node_name][2]
                current_position += 1
        else:
            current_node_name = self.linked_list['tail'][0]
            current_position = 0
            while current_node_name != 'head':
                if current_position == position:
                    self.linked_list[current_node_name][1] = value
                    break
                current_node_name = self.linked_list[current_node_name][0]
                current_position += 1

    # 특정 위치의 데이터 제거하기
    def removeData(self, position, head_or_tail):
        if head_or_tail == 0:
            current_node_name = self.linked_list['head'][2]
            current_position = 0
            while current_node_name != 'tail':
                if current_position == position:
                    prev_node = self.linked_list[current_node_name][0]
                    next_node = self.linked_list[current_node_name][2]
                    self.linked_list[prev_node][2] = next_node
                    self.linked_list[next_node][0] = prev_node
                    del self.linked_list[current_node_name]
                    break
                current_node_name = self.linked_list[current_node_name][2]
                current_position += 1
        else:
            current_node_name = self.linked_list['tail'][0]
            current_position = 0
            while current_node_name != 'head':
                if current_position == position:
                    prev_node = self.linked_list[current_node_name][0]
                    next_node = self.linked_list[current_node_name][2]
                    self.linked_list[prev_node][2] = next_node
                    self.linked_list[next_node][0] = prev_node
                    del self.linked_list[current_node_name]
                    break
                current_node_name = self.linked_list[current_node_name][0]
                current_position += 1
```

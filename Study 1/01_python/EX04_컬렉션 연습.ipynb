학생정보를 출력하는 프로그램을 만든다

학생 정보는 다음과 같이 구성된다.

학생 한명은 이름, 나이, 국어점수, 영어점수, 수학점수로 구성된다.

- 첫 번째 학생 : 홍길동, 15, 90, 91, 92
- 두 번째 학생 : 김길동, 18, 80, 81, 82
- 세 번째 학생 : 최길동, 20, 70, 71, 72

출력은 다음과 같이 한다.

모든 학생들의 정보를 모두 출력한다.

각 점수별 총점과 평균을 출력한다.

전체 총점과 평균을 출력한다.

```python
# 각 학생의 정보를 가지고 있는 딕셔너리를 만들어준다.
studentDict1 = {
    'name' : '홍길동',
    'age' : 15,
    'korean' : 90,
    'english' : 92,
    'math' : 91
}

studentDict2 = {
    'name' : '김길동',
    'age' : 18,
    'korean' : 80,
    'math' : 81,
    'english' : 82
}

studentDict3 = {
    'name' : '최길동',
    'age' : 20,
    'korean' : 70,
    'english' : 72,
    'math' : 71
}
# 딕셔너리를 리스트에 담는다
studentList = [studentDict1, studentDict2, studentDict3]
# 각 점수별 총점과 평균을 구한다
koreanTotal = 0
englistTotal = 0
mathTotal = 0

for d1 in studentList :
    koreanTotal = koreanTotal + d1['korean']
    englishTotal = englistTotal + d1['english']
    mathTotal = mathTotal + d1['math']

koreanAvg = koreanTotal // len(studentList)
englishAvg = englishTotal // len(studentList)
mathAvg = mathTotal // len(studentList)

# 전체 총점과 평균을 구한다.
allTotal = koreanTotal + englishTotal + mathTotal
allAvg = allTotal // len(studentList) // 3
# 모든 학생들의 정보를 출력한다.
for s1 in studentList :
    print(f'이름 : {s1['name']}')
    print(f'나이 : {s1['age']}')
    print(f'국어점수 : {s1['korean']}')
    print(f'영어점수 : {s1['english']}')
    print(f'수학점수 : {s1['math']}')
    print()
# 각 점수별 총점과 평균을 출력한다.
print(f'국어 총점 : {koreanTotal}')
print(f'영어 총점 : {englishTotal}')
print(f'수학 총점 : {mathTotal}')

print(f'국어 평균 : {koreanAvg}')
print(f'영어 평균 : {englishAvg}')
print(f'수학 평균 : {mathAvg}')

# 전체 총점과 평균을 출력한다.
print(f'전체 총점 : {allTotal}')
print(f'전체 평균 : {allAvg}')
```

# 문제1)
자동차의 정보가 다음과 같이 구성되어 있다

자동차 하나의 정보는 다음과 같다

모델명, 연식, 배기량, 색상, 주행거리

첫 번째 자동차 : 붕붕이, 2000, 500, 노란색, 100000
두 번째 자동차 : 방구차, 1950, 200, 빨간색, 200000
세 번째 자동차 : 아스라다, 1980, 5000000, 흰색, 300000

출력은 다음과 같다.

각 자동차의 정보를 모두 출력한다.

총 배기량과 배기량 평균을 출력한다
총 주행거리와 총 주행거리 평균을 출력한다



```python
# 각 자동차의 정보를 가지고 있는 딕셔너리를 만들어준다.
carDict1 = {
    'name' : '붕붕이',
    'age' : 2000,
    'displacement' : 500,
    'color' : '노란색',
    'mileage' : 100000
}

carDict2 = {
    'name' : '방구차',
    'age' : 1950,
    'displacement' : 200,
    'color' : '빨간색',
    'mileage' : 200000
}

carDict3 = {
    'name' : '아스라다',
    'age' : 1980,
    'displacement' : 5000000,
    'color' : '흰색',
    'mileage' : 300000
}
# 딕셔너리를 리스트에 담는다
carList = [carDict1, carDict2, carDict3]

# 각 수치별 총점과 평균을 구한다
displacementTotal = 0
mileageTotal = 0

for d1 in carList :
    displacementTotal = displacementTotal + d1['displacement']
    mileageTotal = mileageTotal + d1['mileage']
   

displacementAvg = displacementTotal // len(carList)
mileageAvg = mileageTotal // len(carList)


# 전체 총점과 평균을 구한다.
allTotal = displacementTotal + mileageTotal 
allAvg = allTotal // len(carList) // 2
# 모든 차들의 정보를 출력한다.
for s1 in carList :
    print(f'이름 : {s1["name"]}')
    print(f'연식 : {s1["age"]}')
    print(f'배기량 : {s1["displacement"]}')
    print(f'색상 : {s1["color"]}')
    print(f'주행거리 : {s1["mileage"]}')
    print()
# 각 수피별 총점과 평균을 출력한다.
print(f'배기량 총점 : {displacementTotal}')
print(f'주행거리 총점 : {mileageTotal}')
print()
print(f'배기량 평균 : {displacementAvg}')
print(f'주행거리 평균 : {mileageAvg}')
```

```python
#강사님 답 (딕셔너리가 아니라 함수로 표현)
# 자동차의 정보를 담는 함수
def setCarInfo(name, year, cc, color, length) :
    car_dict = {
        '이름' : name,
        '연식' : year,
        '배기량' : cc,
        '색상' : color,
        '주행거리' : length
    }
    return car_dict
```

```python
# 총 배기량, 배기량 평균, 총 주행거리, 주행거리 평균을 출력한다.
def show_cc_length_info(total_cc, avg_cc, total_length, avg_length) :
    print(f'총 배기량 : {total_cc}')
    print(f'평균 배기량 : {avg_cc}')
    print(f'총 주행거리 : {total_length}')
    print(f'평균 주행거리 : {avg_length}')
```

```python
# 자동차의 정보를 담는다.
car1 = setCarInfo('붕붕이', 2000, 500, '노란색', 100000)
car2 = setCarInfo('방구차', 1950, 200, '빨간색', 200000)
car3 = setCarInfo('아스라다', 1980, 5000000, '흰색', 300000)
```

```python
# 총 배기량과 배기량 평균을 구해 반환한다.
def get_cc_info(*cars) :
    # 배기량 총합을 담을 변수
    total_cc = 0
    # 자동의 수 만큼 반복한다.
    for car in cars :
        # 배기량을 누적한다
        total_cc = total_cc + car['배기량']

    # 배기량 평균을 구한다.
    avg_cc = total_cc // len(cars)

    return total_cc, avg_cc
```

```python
# 총 배기량과 배기량 평균을 구한다
total_cc, avg_cc = get_cc_info(car1, car2, car3)
#print(total_cc, avg_cc)
```

```python
# 총 주행거리와 주행거리 평균을 구하는 함수
def get_length_info(*cars) :
    # 총 주행거리를 담을 변수
    total_length = 0
    # 자동차의 수 만큼 반복하면서 주행거리를 누적한다.
    for car in cars :
        total_length = total_length + car['주행거리']

    # 주행거리 평균을 구한다.
    avg_length = total_length // len(cars)

    return total_length, avg_length
```

```python
# 총 주행거리와 주행거리 평균을 구한다.
total_length, avg_length = get_length_info(car1, car2, car3)
#print(total_length, avg_length)
```

```python
# 자동차의 정보를 출력한다.
def show_car_info(*cars) :
    for car in cars :
        print(f'이름 : {car["이름"]}')
        print(f'연식 : {car["연식"]}')
        print(f'배기량 : {car["배기량"]}')
        print(f'색상 : {car["색상"]}')
        print(f'주행거리 : {car["주행거리"]}')
        print()
```

```python
# 자동차의 정보를 출력한다.
show_car_info(car1, car2, car3)

# 총 배기량, 배기량 평균, 총 주행거리, 주행거리 평균을 출력한다.
show_cc_length_info(total_cc, avg_cc, total_length, avg_length)
```

모든 동물의 정보를 출력해주세요

부가적으로 어떠한 정보를 출력할지 자유롭게 결정해주세요

```python
# 각 동물의 정보를 가지고 있는 딕셔너리를 만들어준다.
animalDict1 = {
    'name' : '호랑이',
    'age' : 15,
    'weight' : 200,
    'length' : 240,
}

animalDict2 = {
    'name' : '사자',
    'age' : 5,
    'weight' : 170,
    'length' : 0,
}

animalDict3 = {
    'name' : '늑대',
    'age' : 8,
    'weight' : 80,
    'length' : 170,
}
# 딕셔너리를 리스트에 담는다
animalList = [animalDict1, animalDict2, animalDict3]
# 각 수치 별 총점과 평균을 구한다
ageTotal = 0
weightTotal = 0
lengthTotal = 0

for d1 in animalList :
    ageTotal = ageTotal + d1['age']
    weightTotal = weightTotal + d1['weight']
    lengthTotal = lengthTotal + d1['length']

ageAvg = ageTotal / len(animalList)
weightAvg = weightTotal / len(animalList)
lengthAvg = lengthTotal / len(animalList)

# 전체 수치의 총 합과 평균을 구한다.
allTotal = ageTotal + weightTotal + lengthTotal
allAvg = allTotal / len(animalList) / 3
# 모든 동물들의 정보를 출력한다.
for s1 in animalList :
    print(f'이름 : {s1['name']}')
    print(f'나이 : {s1['age']}')
    print(f'무게 : {s1['weight']}')
    print(f'길이 : {s1['length']}')
    print()
# 각 동물별 수치의 총 합과 평균을 출력한다.
print(f'나이 총합 : {ageTotal}')
print(f'무게 총합 : {weightTotal}')
print(f'길이 총합 : {lengthTotal}')
print()

print(f'나이 평균 : {ageAvg}')
print(f'무게 평균 : {weightAvg}')
print(f'길이 평균 : {lengthAvg}')
```
